```python
# Encoding

from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()

df['Vehicle Type']=lb.fit_transform(df['Vehicle Type'])

df['Weather']=lb.fit_transform(df['Weather'])

df['Economic Condition']=lb.fit_transform(df['Economic Condition'])

df['Day Of Week']=lb.fit_transform(df['Day Of Week'])

df['Is Peak Hour']=lb.fit_transform(df['Is Peak Hour'])

df['Random Event Occurred']=lb.fit_transform(df['Random Event Occurred'])

df['Traffic Density']=lb.fit_transform(df['Traffic Density'])

df.dtypes
```

```python
df.drop(['City'],axis=1,inplace=True)
df
```

```python
# Splittin Input and Output Data
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
x
```

```python
y
```

```python
# Splitting Training and Testing data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#MinMax Scaler Normalization

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
# Model Creation

from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
sv=SVC()
sv.fit(x_train,y_train)
y_pred=sv.predict(x_test)
y_pred
```

```python
y_train
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
report=classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
labels=['1','2','0']
cmd=ConfusionMatrixDisplay(cm,display_labels=labels)
cmd.plot()
```
